```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig,pipeline
```

```python
HF_TOKEN="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" # Replace with your actual Hugging Face token
model_name="meta-llama/Meta-Llama-3-8B"

bnb_config=BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

# Pass the HF_token as the 'use_auth_token' parameter
tokenizer= AutoTokenizer.from_pretrained(model_name, token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token
```

```python
model= AutoModelForCausalLM.from_pretrained(model_name,
                                            quantization_config=bnb_config,
                                            token=HF_TOKEN,
                                            device_map="auto") # Keep only 'type' and 'factor' keys in rope_scaling
```

```python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
```

```python
def get_response(prompt):
  response = text_generator(prompt)
  return response[0]['generated_text']
```

```python
prompt="job description for generative ai"
print(get_response(prompt))
```
